<h2>Modeling</h2>
<h4>Problem Statement</h4><p><em>Model the values for 'Close' for the entire dataset</em></p>

<h4>The methodology</h4>
<ul>
    <li>We will drop column Adjusted Close</li>
    <li>Out of remaining variables, we keep variable Open since it is the most correlated with the target variable Close. We drop others</li>
</ul>

<h4>Approach</h4>
<ul>
    <li>We will use a simple RNN with each timestep corrsponding to each data point. We dont need Date column and as such has been dropped. If need be, we can use Date list created earlier after formating the str to datetime</li>
    <li>We will make the data stationary and see if that makes a difference.</li>

</ul>

```python
#We also drop Date. We will use Date variable of type list of datetime created earlier, if need be.
df.drop(columns=['High', 'Low', 'Adj Close', 'Date'], inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler(feature_range=(-1,1))
Scaler.fit(df)
dset = Scaler.transform(df)
```

```python
#--Split data into train and test sets

#90% of 4598 = 4138
train = dset[:4138, :]
test = dset[:(4598-4138), :]
Date_test = Date[:(4598-4138)]

#--Split into X and Y
X_train = train[:, :-1]
Y_train = train[:, 1:]

X_test = test[:, :-1]
Y_test = test[:, 1:]

#Note that the above sets will not be modified further. 
#Any further processing will be done on copies of the above
#and only on Train sets.
```

<h3>Model 1</h3>
<p>We will build a stacked LSTM where we split the data into smaller sequences, say 120 timesteps each. For example, one sequence may contain 120 consequtive values of column Open and we have to predict the value of column Close that corresponds to 120th value of the input sequence.</p>

<p>Input X_Mn will be a 3D tensor with shape (samples, timesteps, features). We have a sequence of 4138 values. We use a sliding window of 120 values to build batches. Hence the shape becomes ((4138-120), 120, 1) </p>

```python
#convert X into 3D tensor. Subsampling
def Subsample(X, Y, window=60):
    xtrain = []
    ytrain = []
    for i in range(X.shape[0]-window):
        for j in range(window):
            xtrain.append(X[j+i])

    ytrain = Y[window:]

    x = np.array(xtrain, dtype=list).reshape(X.shape[0]-window, window, 1)
    y = np.array(ytrain)
    
    return x, y
```

```python
X_M1, Y_M1 = Subsample(X_train, Y_train)

X_M1.shape, Y_M1.shape
```

```python
import tensorflow as tf
import tensorflow.keras as keras
from keras.layers import LSTM, Input, Dense
from keras import Model
```

```python
dense = Dense(1, activation=None)

inp = Input(shape=(60,1))
x = LSTM(units=50, return_sequences=True, dropout=0.2)(inp)
x = LSTM(units=50, return_sequences=True, dropout=0.2)(x)
x = LSTM(units=50, return_sequences=True, dropout=0.2)(x)
x = LSTM(units=50, return_sequences=False, dropout=0.2)(x)
out = dense(x)

M1 = Model(inp, out)
```

```python
M1.compile('adam', 'mse')
hist = M1.fit(x=X_M1.astype('float32'), y=Y_M1, epochs=15, validation_split=0.1)
```

```python
Loss = {'M1' : hist.history['val_loss'], 'M1-Final' : hist.history['val_loss'][-1]}
```

<h3>Model 2</h3>
<p>This time we use the same model 1 but we preprocess the data with lag=1 differencing.</p>

```python
def difference(X, Y, lag = 1):
    xlag = np.array([X[i+lag] - X[i] for i in range(X.shape[0]-1)])
    ylag = np.array([Y[i+lag] - Y[i] for i in range(Y.shape[0]-1)])
    return xlag, ylag

def inverse_difference(xlag, ylag, X, Y, lag = 1):
    invX = np.array([xlag[i]+X[i] for i in range(xlag.shape[0])])
    invY = np.array([ylag[i]+Y[i] for i in range(ylag.shape[0])])
    return invX, invY
```

```python
X_M2, Y_M2 = difference(X_train, Y_train)

X_M2, Y_M2 = Subsample(X_M2, Y_M2, window=60)

X_M2.shape, Y_M2.shape
```

```python
inp = Input(shape=(X_M2.shape[1], X_M2.shape[2]))
x = LSTM(units=50, return_sequences=True, dropout=0.2)(inp)
x = LSTM(units=50, return_sequences=True, dropout=0.2)(x)
x = LSTM(units=50, return_sequences=True, dropout=0.2)(x)
x = LSTM(units=50, return_sequences=False, dropout=0.2)(x)
out = dense(x)

M2 = Model(inp, out)
```

```python
M2.compile('adam', 'mse')
hist = M2.fit(x=X_M2.astype('float32'), y=Y_M2.astype('float32'), epochs=15, validation_split=0.1)
```

```python
Loss['M2'] = hist.history['val_loss']
Loss['M2-Final'] = hist.history['val_loss'][-1]
```

<h2>Validation</h2>
<p>We have preliminary results from 2 models and seems like M2 (M1 with lag=1 differencing) performs better.</p>

```python
Loss['M1-Final'], Loss['M2-Final']
```

```python
fig = plt.figure(figsize=(20, 5))

ax = fig.add_subplot(121)
ax.plot(Loss['M2'])
ax.set_title('Model 2 val_loss')

ax = fig.add_subplot(122)
ax.plot(Loss['M1'])
ax.set_title('Model 1 val_loss')

plt.show()
```

```python
#prediction on test data

#P
Open, Close = X_test, Y_test
Open, _ = Subsample(Open, Close)
Yhat = M1.predict(Open.astype('float32'))

#Inverse Process
ly = pd.DataFrame()
ly['Yhat'] = Yhat[:, 0]
ly['Y'] = Close[:-60]
to_plot = Scaler.inverse_transform(ly)

#Plot
plt.plot(to_plot[:, 0], 'r', label = 'Yhat')
plt.plot(to_plot[:, 1], 'g', label = 'Y')
plt.legend()
plt.show()
```

```python
#Model Complete dataset for Model 1
#Preprocess whole dset
Open, Close = dset[:, 0], dset[:, 1]
Open, _ = Subsample(Open, Close)
Yhat = M1.predict(Open.astype('float32'))

#Inverse Process
ls = pd.DataFrame()
ls['Yhat'] = Yhat[:, 0]
ls['Y'] = Close[:-60]
to_plot = Scaler.inverse_transform(ls)

#Plot
plt.plot(to_plot[:, 0], 'r', label = 'Yhat')
plt.plot(to_plot[:, 1], 'g', label = 'Y')
plt.legend()
plt.show()
```

```python
#Preprocess whole dset
Open, Close = dset[:, 0], dset[:, 1]
Open, _ = Subsample(Open, Close)
Open, _ = difference(Open, _)
Yhat = M2.predict(Open.astype('float32'))

#Inverse Process
Yhat, _ = inverse_difference(Yhat[:, 0], _, Close[:-60], _)
ls = pd.DataFrame()
ls['Yhat'] = Yhat
ls['Y'] = Close[:-61]
to_plot = Scaler.inverse_transform(ls)

#Plot
plt.plot(to_plot[:, 0], 'r', label = 'Yhat')
plt.plot(to_plot[:, 1], 'g', label = 'Y')
plt.legend()
plt.show()
```
